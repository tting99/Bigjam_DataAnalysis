```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import skew

# ECLO 값이 있는 train 데이터 불러오기
train_df = pd.read_csv('./open/train.csv', encoding='cp949', low_memory=False)
```

## **ECLO 왜도 확인**   

3.35로 비교적 높은 왜도 값 확인, 로그 변환이 필요함

```python
# skew 함수 이용해 왜도 확인
eclo_skewness = skew(train_df['ECLO'])

# 결과 출력
print(eclo_skewness)
```

### **왼쪽으로 치우져져 있는 결과 확인**

```python
train_df['ECLO'].hist()
plt.show()
```

## **최적의 로그변환 함수 알아보기**   

- `log`, `log1p`, `log2` 각 함수의 로그변환 시 ECLO 점수 변환을 시각화해 비교
- 각 로그변환에 대한 왜도값을 비교 후 최종 로그변환 방식 설정
- **왜도값이 가장 낮은 `log2` 로그 변환 방식 채택**

```python
org_skew = skew(train_df['ECLO'])
log_skew = skew(np.log(train_df['ECLO']))
log1p_skew = skew(np.log1p(train_df['ECLO']))
log2_skew = skew(np.log2(train_df['ECLO']))

# 각 변환을 적용한 후의 분포 시각화
plt.figure(figsize=(12, 6))

# Original
plt.subplot(2, 2, 1)
plt.hist(train_df['ECLO'], bins=20, edgecolor='black')
plt.title('Original ECLO Scores')

# np.log 변환
plt.subplot(2, 2, 2)
plt.hist(np.log(train_df['ECLO']), bins=20, edgecolor='black')
plt.title('Log-transformed (np.log) ECLO Scores')

# np.log1p 변환
plt.subplot(2, 2, 3)
plt.hist(np.log1p(train_df['ECLO']), bins=20, edgecolor='black')
plt.title('Log-transformed (np.log1p) ECLO Scores')

# np.log1p 변환
plt.subplot(2, 2, 4)
plt.hist(np.log2(train_df['ECLO']), bins=20, edgecolor='black')
plt.title('Log-transformed (np.log2) ECLO Scores')

plt.tight_layout()
plt.show()

print('로그변환 방식별 왜도')
print('original ECLO skew : ', org_skew)
print('log ECLO skew : ', log_skew)
print('log1p ECLO skew : ', log1p_skew)
print('log2 ECLO skew : ', log2_skew)
```
